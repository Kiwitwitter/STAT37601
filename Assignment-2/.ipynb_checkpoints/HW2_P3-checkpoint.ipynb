```python
import pickle
f=open('project/cmsc25025/sou/speeches.pkl','rb')
speeches=pickle.load(f)
import numpy as np
import matplotlib.pyplot as plt
import string
import re
```

#### (a)

```python
lens=len(speeches)
spenew = list()
for i in range(lens):
    sentence = speeches[i].text
    spenew.append(sentence)
```

```python
def clean_and_split(s):
    # convert to lowercase and translate all hyphens and
    # punctuation to whitespace
    translator = str.maketrans(dict.fromkeys(string.punctuation))
    s = s.replace("-"," ").translate(translator).lower()
    # replace \r\n and \x95
    s = re.sub("(\r\n)+"," ",s)
    s = re.sub("(\x95)+"," ",s)
    # replace whitespace substrings with one whitespace and remove
    # leading/trailing whitespaces
    s = re.sub(" +"," ",s.strip())
    return s.split(" ")
```

```python
cleanspe = list()
for i in range(lens):
    clean = clean_and_split(spenew[i])
    cleanspe.append(clean)
```

```python
count = list()
for i in range(lens):
    wordcount = dict()
    for word in cleanspe[i]:
        if word in wordcount.keys():
            wordcount[word]+=1
        else:
            wordcount[word]=1
    count.append(wordcount)
```

```python
voc = dict()
for i in range(len(count)):
    for word in count[i].keys():
        if word in voc.keys():
            voc[word]= voc[word]+count[i][word]
        else:
            voc[word]= count[i][word]
```

```python
voca=sorted(voc.items(), key=lambda e:e[1], reverse=True)
```

```python
newcount = voca[20:]
```

```python
finalcount = list()
for i in range(len(newcount)):
    if newcount[i][1] >= 50:
        finalcount.append(newcount[i])
```

```python
len(finalcount)
```

```python
total = list()
for i in range(lens):
    dic = dict()
    for j in range(len(finalcount)):
        if finalcount[j][0] in count[i].keys():
            dic[finalcount[j][0]]= count[i][finalcount[j][0]]
        else:
            dic[finalcount[j][0]]= 0
    total.append(dic)           
```

```python
total[0]
```

```python
vocabu = list()
for word in total[0].keys():
    vocabu.append(word)
```

```python
voclist = sorted(vocabu)
```

```python
totlist = list()
for i in range(len(total)):
    list1 = list()
    for j in range(len(voclist)):
        if voclist[j] in total[i].keys():
            list1.append(total[i][voclist[j]])
        else:
            list1.append(0)
    totlist.append(list1)
```

```python
di = np.zeros(len(totlist[0]))
for j in range(len(totlist[0])):
    for i in range(len(totlist)):
        if totlist[i][j] != 0:
            di[j] = di[j]+1
        else:
            di[j] = di[j]
di=di.tolist()
```

```python
wtotal = list()
for i in range(lens):
    w = np.zeros(len(totlist[0]))
    for j in range(len(totlist[0])):
        w[j] = totlist[i][j]*np.log(lens/di[j])
    wtotal.append(w)
```

```python
from numpy import linalg as LA
similar = list()
for i in range(len(wtotal)):
    simila = np.zeros(len(wtotal))
    for j in range(len(wtotal)):
        simi = np.dot(wtotal[i],wtotal[j])/(LA.norm(wtotal[i])*LA.norm(wtotal[j]))
        simila[j] = simi
    similar.append(simila)   
```

```python
# Store president's name of each SOU in a list 
president = list()
for i in range(lens):
    president.append(speeches[i].president)
```

```python
sim_diff = list()
for i in range(len(wtotal)):
    for j in range(len(wtotal)):
        if president[i] != president[j] and j > i:
            a = np.array([similar[i][j],i,j])
            sim_diff.append(a)
```

```python
top50_diff = sorted(sim_diff,key=lambda x: x[0],reverse=True)[:50]
top50_diff
```

```python
pair_diff = list()
for i in range(50):
    pre = np.array([president[np.int(top50_diff[i][1])],president[np.int(top50_diff[i][2])]])
    pair_diff.append(pre)
```

#### (b)

```python
sim_same = list()
for i in range(len(wtotal)):
    for j in range(len(wtotal)):
        if president[i] == president[j] and i != j and j > i:
            b = np.array([similar[i][j],i,j])
            sim_same.append(b)
```

```python
top50_same = sorted(sim_same,key=lambda x: x[0],reverse=True)[:50]
top50_same
```

```python
pair_same = list()
for i in range(50):
    pre = president[np.int(top50_same[i][1])]
    pair_same.append(pre)
```

```python
pair_same
```

```python
# Store presidents' names and corresponding total number of words and sentences in two dictionaries
pres_vec = dict()
for i in range(lens):
    pres_name = president[i]
    if pres_name in pres_vec.keys():
        pres_vec[pres_name] = np.vstack([pres_vec[pres_name], wtotal[i]])
    else:
        pres_vec[pres_name] = wtotal[i]
```

```python
pres_veclis = sorted(pres_vec.items(), key=lambda e:e[0])
```
